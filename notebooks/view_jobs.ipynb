# Job postings viewer

Use this notebook to inspect the saved Delta tables in VS Code without relying on terminal output.

Start by running the cells below. They read the silver job table and then summarize the gold tables.

```python
from pathlib import Path

from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("view-jobs").getOrCreate()
lake_root = Path("data/lake")
silver_path = lake_root / "silver" / "jobs"
silver = spark.read.format("delta").load(str(silver_path))
silver.select(
    "title",
    "company",
    "country_code",
    "city",
    "seniority",
    "skills",
    "salary_min",
    "salary_max",
).show(20, truncate=False)
```

```python
gold_root = lake_root / "gold"
gold_tables = [
    "skills_by_country",
    "top_tools",
    "cloud_demand",
    "jobs_by_day",
    "top_companies",
]

for table in gold_tables:
    df = spark.read.format("delta").load(str(gold_root / table))
    print(f"{table}: {df.count()} rows")
    df.show(5, truncate=False)
```
